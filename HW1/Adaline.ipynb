```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

# Generate the data as before
np.random.seed(0)
blue_x = np.random.normal(loc=0, scale=0.1, size=100)
blue_y = np.random.normal(loc=0, scale=0.4, size=100)
blue_labels = np.array([0] * 100)  # Assign label 0 to Blue

np.random.seed(1)
red_x = np.random.normal(loc=1, scale=0.2, size=100)
red_y = np.random.normal(loc=1, scale=0.2, size=100)
red_labels = np.array([1] * 100)  # Assign label 1 to Red

# Combine the data
x_data = np.concatenate((blue_x, red_x))
y_data = np.concatenate((blue_y, red_y))
labels = np.concatenate((blue_labels, red_labels))

# Convert data to PyTorch tensors
x_data = torch.FloatTensor(x_data)
y_data = torch.FloatTensor(y_data)
labels = torch.LongTensor(labels)

# Define the Adaline model
class Adaline(nn.Module):
    def __init__(self, input_size, output_size):
        super(Adaline, self).__init__()
        self.fc = nn.Linear(input_size, output_size)

    def forward(self, x):
        return self.fc(x)

# Create the Adaline model
input_size = 2
output_size = 2
adaline = Adaline(input_size, output_size)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(adaline.parameters(), lr=0.01)

# Lists to store loss and accuracy values during training
loss_history = []
accuracy_history = []

# Train the Adaline model
epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = adaline(torch.stack((x_data, y_data), dim=1))
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()

    # Calculate accuracy
    with torch.no_grad():
        predicted = torch.argmax(outputs, dim=1)
        accuracy = (predicted == labels).sum().item() / len(labels)

    # Append loss and accuracy to history lists
    loss_history.append(loss.item())
    accuracy_history.append(accuracy)

    if (epoch + 1) % 100 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Loss: {loss.item():.4f}, Accuracy: {accuracy * 100:.2f}%')

# Plot loss and accuracy
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(range(epochs), loss_history)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Over Time')

plt.subplot(1, 2, 2)
plt.plot(range(epochs), accuracy_history)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy Over Time')

plt.tight_layout()
plt.show()

# Visualize the decision boundary
plot_decision_boundary(adaline, x_data, y_data, labels)
```
